```python
import re

Label_file='./dev_set/Query_Label'
Query_result ='./66.txt'
a = []
b = []
with open(Query_result) as fq:
    for line in fq.readlines():
        line = re.sub('\n','',line)
        line = re.split("(,|\[|\]|\s)", line)
        temp = []
        for ele in line:
            if ele.isdigit():
                temp.append(int(ele))
        a.append(temp)
with open(Label_file) as fl:
    for line in fl.readlines():
        line = re.sub('\n','',line)
        line = line.split(' ')
        intline = [int(i) for i in line]
        b.append(intline)

sum = len(a)
total = 0
for i in range(sum):
    for j in range(len(a[i])):
        if a[i][j] != b[i][j]:
            total += 1
    print('line', i,'result:',a[i], 'label:',b[i])
print('total:',total)
```
